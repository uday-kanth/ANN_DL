Indexing And Slicing

```python
import tensorflow as tf
tf.__version__
```

Indexing

```python
x=tf.random.normal([4,32,32,3])
x[0]
```

```python
x[0][1]
```

```python
x[0][1][2]
```

```python
x[0][1][2][0]
```

Slicing

```python
x[1:3]
```

```python
x[0,::]
```

```python
x[:,0:28:2,0:28:2,:]
```

```python
x=tf.range(9)
x[8:0:-1]
```

```python
x[::-1]
```

```python
x[::-2]
```

```python
x=tf.random.normal([4,32,32,3])
x[0,::-2,::-2]
```

```python
x[0:2,...,1:]
```

```python
x[2:,...] # same as x[2:]
```

```python
x[...,:2]
```

Mathematical Operations

```python
a=tf.range(5)
b=tf.constant(2)
```

```python
a+b
```

```python
a-b
```

```python
x=tf.range(4)
tf.pow(x, 3)
```

```python
x=tf.constant([1.,4.,9.])
x**(0.5)
```

```python
x=tf.range(5)
x=tf.cast(x, dtype=tf.float32)
x=tf.square(x)
x
```

```python
a=tf.random.normal([4,3,28,32])
b=tf.random.normal([4,3,32,3])
a@b # @ is matrix multiplication
```

```python
a=tf.random.normal([4,28,32])
b=tf.random.normal([32,16])
tf.matmul(a, b)
```
